# 🛳 Titanic Survival Prediction - Project Summary

In this project, we built a machine learning model to predict whether a passenger on the Titanic would survive or not, based on features like age, gender, class, number of siblings/spouses aboard, etc.

---

## 📌 Steps Taken

### 1. Exploratory Data Analysis (EDA):
We explored the dataset to understand the distribution of passengers, survival rates, and patterns within features like **Sex**, **Pclass**, and **Age**.

### 2. Data Cleaning & Preprocessing:
We handled missing values, encoded categorical variables, and dropped unnecessary columns like `PassengerId`. This helped make the dataset suitable for machine learning models.

---

### 3. Modeling:
We trained a **Random Forest Classifier**, which gave us an **accuracy of about 82%**.
